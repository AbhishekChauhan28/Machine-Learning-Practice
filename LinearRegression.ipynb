```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
df = datasets.load_diabetes()
```

```python
df.keys()
```

```python
df.DESCR
```

```python
df_X = df.data[:, np.newaxis, 2]
df_X
```

```python
regression = linear_model.LinearRegression()
```

```python
df_X_train = df_X[:-250]
df_X_test = df_X[100:]
                
df_Y_train = df.target[:-250]
df_Y_test = df.target[100:]
```

```python
regression.fit(df_X_train, df_Y_train)
```

```python
df_Y_predicted = regression.predict(df_X_test)
```

```python
print("Mean Squared error is : ", mean_squared_error(df_Y_test, df_Y_predicted))
```

```python
print("weights : ", regression.coef_)
print("Intercept : ", regression.intercept_)
```

```python
plt.scatter(df_X_test, df_Y_test)
plt.plot(df_X_test, df_Y_predicted)
plt.show()
```
